```python
import sqlite3
import pandas as pd
import os

# Create a new SQLite database (it will be saved as a file)
conn = sqlite3.connect("music_store.db")
print("‚úÖ SQLite database created/connected successfully.")
```

```python
employee = pd.read_csv("employee.csv")
customer = pd.read_csv("customer.csv")
invoice = pd.read_csv("invoice.csv")
invoice_line = pd.read_csv("invoice_line.csv")
track = pd.read_csv("track.csv")
album = pd.read_csv("album.csv")
artist = pd.read_csv("artist.csv")
genre = pd.read_csv("genre.csv")
media_type = pd.read_csv("media_type.csv")
playlist = pd.read_csv("playlist.csv")
playlist_track = pd.read_csv("playlist_track.csv")

print("‚úÖ All CSVs loaded into Pandas successfully.")
```

Q1 ‚Äì Most senior employee

```python
pd.read_sql_query("""
SELECT first_name, last_name, title
FROM employee
ORDER BY levels DESC
LIMIT 1;
""", conn)
```

Q2 ‚Äì Countries with most invoices

```python
pd.read_sql_query("""
SELECT billing_country, COUNT(*) AS invoice_count
FROM invoice
GROUP BY billing_country
ORDER BY invoice_count DESC;
""", conn)
```

Q3 ‚Äì Top 3 invoice totals

```python
pd.read_sql_query("""
SELECT invoice_id, total
FROM invoice
ORDER BY total DESC
LIMIT 3;
""", conn)
```

Q4 ‚Äì City with highest total invoice amount

```python
pd.read_sql_query("""
SELECT billing_city, SUM(total) AS total_sales
FROM invoice
GROUP BY billing_city
ORDER BY total_sales DESC
LIMIT 1;
""", conn)
```

Q5 ‚Äì Customer who spent the most

```python
pd.read_sql_query("""
SELECT c.first_name, c.last_name, SUM(i.total) AS total_spent
FROM customer c
JOIN invoice i ON c.customer_id = i.customer_id
GROUP BY c.customer_id
ORDER BY total_spent DESC
LIMIT 1;
""", conn)
```

6Ô∏è‚É£ Moderate Level Queries

Q1 ‚Äì Customers who listen to Rock music

```python
pd.read_sql_query("""
SELECT DISTINCT c.email, c.first_name, c.last_name
FROM customer c
JOIN invoice i ON c.customer_id = i.customer_id
JOIN invoice_line il ON i.invoice_id = il.invoice_id
JOIN track t ON il.track_id = t.track_id
JOIN genre g ON t.genre_id = g.genre_id
WHERE g.name = 'Rock'
ORDER BY c.last_name;
""", conn)
```

Q2 ‚Äì Top 10 Rock artists by track count

```python
pd.read_sql_query("""
SELECT ar.name AS artist_name, COUNT(t.track_id) AS track_count
FROM artist ar
JOIN album al ON ar.artist_id = al.artist_id
JOIN track t ON al.album_id = t.album_id
JOIN genre g ON t.genre_id = g.genre_id
WHERE g.name = 'Rock'
GROUP BY ar.artist_id
ORDER BY track_count DESC
LIMIT 10;
""", conn)
```

Q3 ‚Äì Tracks longer than average

```python
pd.read_sql_query("""
SELECT name, milliseconds
FROM track
WHERE milliseconds > (SELECT AVG(milliseconds) FROM track)
ORDER BY milliseconds DESC;
""", conn)
```

Q1 ‚Äì Each customer‚Äôs spending per artist

```python
pd.read_sql_query("""
WITH artist_sales AS (
    SELECT ar.artist_id, ar.name AS artist_name, il.invoice_id,
           SUM(il.unit_price * il.quantity) AS artist_revenue
    FROM invoice_line il
    JOIN track t ON il.track_id = t.track_id
    JOIN album al ON t.album_id = al.album_id
    JOIN artist ar ON al.artist_id = ar.artist_id
    GROUP BY ar.artist_id, il.invoice_id
)
SELECT c.first_name, c.last_name, a.artist_name, SUM(a.artist_revenue) AS total_spent
FROM customer c
JOIN invoice i ON c.customer_id = i.customer_id
JOIN artist_sales a ON i.invoice_id = a.invoice_id
GROUP BY c.customer_id, a.artist_id
ORDER BY total_spent DESC;
""", conn)
```

Q2 ‚Äì Most popular genre per country

```python
pd.read_sql_query("""
WITH genre_country AS (
    SELECT c.country, g.name AS genre_name, COUNT(il.invoice_line_id) AS purchase_count
    FROM invoice_line il
    JOIN invoice i ON il.invoice_id = i.invoice_id
    JOIN customer c ON i.customer_id = c.customer_id
    JOIN track t ON il.track_id = t.track_id
    JOIN genre g ON t.genre_id = g.genre_id
    GROUP BY c.country, g.genre_id
),
ranked AS (
    SELECT country, genre_name, purchase_count,
           RANK() OVER (PARTITION BY country ORDER BY purchase_count DESC) AS rank
    FROM genre_country
)
SELECT country, genre_name, purchase_count
FROM ranked
WHERE rank = 1;
""", conn)
```

Q3 ‚Äì Top-spending customer per country

```python
pd.read_sql_query("""
WITH customer_spending AS (
    SELECT c.country, c.customer_id, c.first_name, c.last_name, SUM(i.total) AS total_spent
    FROM customer c
    JOIN invoice i ON c.customer_id = i.customer_id
    GROUP BY c.country, c.customer_id
),
ranked AS (
    SELECT *, RANK() OVER (PARTITION BY country ORDER BY total_spent DESC) AS rank
    FROM customer_spending
)
SELECT country, first_name, last_name, total_spent
FROM ranked
WHERE rank = 1;
""", conn)
```
